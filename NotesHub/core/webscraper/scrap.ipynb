```python
import requests
import html5lib
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
import re
```

```python
driver = webdriver.Chrome()


url = "https://www.cse.iitb.ac.in/academics/courses"
driver.get(url)


driver.maximize_window()

try:
    wait = WebDriverWait(driver, 10)
    autumn = driver.find_element(By.CLASS_NAME, "accordion-button.collapsed")
    autumn.click()
    content = driver.page_source
    soup = BeautifulSoup(content, "html.parser")

except Exception as e:
    print(f"An error occurred: {e}")

driver.quit()
```

```python
table = content.split("<span title=")
course_names = []
table = table[1:len(table)]
i = 0
for names in table:
    name = names.split("class=")
    if(name[0].find("Prof") == -1):
        course_names.append(name[0][1:len(name[0]) - 2])
    
```

```python
table = content.split("course-id ng-star-inserted")
table = table[1:]
code = []
for codes in table:
    codes = codes.split(" </a>")
    codes = codes[0].split("> ")
    code.append(codes[1])
```

```python
short_name = []
for name in course_names:
    a = ""
    for i in name:
        if(i <= 'Z' and i >= 'A'):
            a += i
    short_name.append(a)
```

```python
df = pd.DataFrame()
df['Course Code'] = code
df['Course Name'] = course_names
df['Short names'] = short_name
df.to_csv("courses.csv",index = False)
df
```
